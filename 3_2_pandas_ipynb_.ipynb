```python
def edu_groups(series): # Series : 한 칼럼을 갖고온다
    if series==2:
        return 1
    elif series==3 or series==4:
        return 2
    elif series>=5:
        return 3
```

```python
df['eduM']=df['edu'].apply(edu_groups) # edu에 함수를 적용한다
df.head()
```
